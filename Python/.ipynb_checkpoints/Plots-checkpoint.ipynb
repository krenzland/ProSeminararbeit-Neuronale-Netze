```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

# testing changing sigmoid width (and slope at the same time) - parameter b
x = np.linspace(-6,6,256)
y = sigmoid(x) # default
y2 = 1.7159 * np.tanh( 0.66666 * x)
y3 = np.fmax(0,x)
# Create the plot
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x,y, lw=2,color='black', label="Logistic")
plt.plot(x,y2, lw=2,color='blue', label="Tanh")
#plt.plot(x,y3, lw=2,color='red', label="Rectifier")


plt.legend(loc='upper left')
savefig('Aktivierungsfunktionen.pdf', transparent=True)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sin((1/2 * X**2) - 1/4*Y**2 + 3) * np.cos(2*X + 1 - np.exp(Y)) #np.sqrt(X**2 + Y**2)
Z = X + Y #np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

#fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
